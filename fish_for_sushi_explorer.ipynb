# Fish for Sushi

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fish_for_sushi = pd.read_csv('datasets/fish_for_sushi.csv')
fish_for_sushi.head()
```

```python
#sum each Product Name by Volume (kg) for each year
fish_for_sushi_grouped = fish_for_sushi.groupby(['Year', 'Product Name'])['Volume (kg)'].sum().reset_index()
fish_for_sushi_grouped.head()


#Line graph: volume of product in kg by year for each product name, exclude 2026
plt.figure(figsize=(12, 6))
for product in fish_for_sushi_grouped['Product Name'].unique():
    product_data = fish_for_sushi_grouped[fish_for_sushi_grouped['Product Name'] == product]
    product_data = product_data[product_data['Year'] < 2026]  # Exclude 2026
    plt.plot(product_data['Year'], product_data['Volume (kg)'], label=product)
plt.title('Volume of Fish imported to USA per year')
plt.xlabel('Year')
plt.ylabel('Volume (kg)')
plt.legend()
```
